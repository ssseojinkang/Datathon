```python
import requests
from datetime import datetime
import pandas as pd
import numpy as np
```

```python
# 형식 예시 fromdata = '2021-03-21'
# 형식 예시 todate = '2021-03-21'
def get_game_id(fromdate, todate):
    url = f"https://api-gw.sports.naver.com/schedule/games?fields=basic%2CsuperCategoryId%2CcategoryName%2Cstadium%2CstatusNum%2CgameOnAir%2ChasVideo%2Ctitle%2CspecialMatchInfo%2CroundCode%2CseriesOutcome%2CseriesGameNo%2ChomeStarterName%2CawayStarterName%2CwinPitcherName%2ClosePitcherName%2ChomeCurrentPitcherName%2CawayCurrentPitcherName%2CbroadChannel&upperCategoryId=kbaseball&categoryId=kbo&fromDate={fromdate}&toDate={todate}&roundCodes=&size=500"
    
    response = requests.get(url)
    json_data = response.json().get("result", {}).get("games", {})
    
    # 바로 리스트 반환
    return [f"gameid{i+1}: {game['gameId']}" for i, game in enumerate(json_data)]
```

```python
# 데일리 게임 정보 불러오기

## 선발 이름 정보 같이 추가해야함
def progress_game_info(game_id):
    # url = f"https://api-gw.sports.naver.com/schedule/games/{game_id}/record"
    url = f"https://api-gw.sports.naver.com/schedule/games/{game_id}/record"

    response = requests.get(url)
    json_data = response.json().get("result", {}).get("recordData", {})

    if json_data:
        game_info = json_data.get("gameInfo", {})
        score_board = json_data.get("scoreBoard", {}).get("rheb", {})
        key_stats = json_data.get("todayKeyStats", {})
        pitchers_boxscore = json_data.get("pitchersBoxscore", {})
        game_date_time = f"{game_info.get('gdate')} {game_info.get('gtime')}"
        game_date_time = datetime.strptime(game_date_time, '%Y%m%d %H:%M')
        game_date_time = game_date_time.strftime('%Y-%m-%d %H:%M:%S')
        home_starter = pitchers_boxscore.get('home', [{}])[0].get('name', None) if pitchers_boxscore.get('home') else None
        away_starter = pitchers_boxscore.get('away', [{}])[0].get('name', None) if pitchers_boxscore.get('away') else None

        
        return {
            "game_id": game_id,
            "game_date_time": game_date_time,
            "home_team": game_info.get("hName"),
            "away_team": game_info.get("aName"),
            "home_score": score_board.get("home", {}).get("r", 0),
            "away_score": score_board.get("away", {}).get("r", 0),
            "score_difference": abs(score_board.get("home", {}).get("r", 0) - score_board.get("away", {}).get("r", 0)),
            "home_hits": float(key_stats.get("home", {}).get("hit", 0)),
            "home_bb": float(score_board.get("home", {}).get("b", 0)),
            "home_homeruns": float(key_stats.get("home", {}).get("hr", 0)),
            "home_stolenbases": float(key_stats.get("home", {}).get("sb", 0)),
            "home_strikeouts": float(key_stats.get("home", {}).get("kk", 0)),
            "home_doubleplays": float(key_stats.get("home", {}).get("gd", 0)),
            "home_errors": float(key_stats.get("home", {}).get("err", 0)),
            "away_hits": float(key_stats.get("away", {}).get("hit", 0)),
            "away_bb": float(score_board.get("away", {}).get("b", 0)),
            "away_homeruns": float(key_stats.get("away", {}).get("hr", 0)),
            "away_stolenbases": float(key_stats.get("away", {}).get("sb", 0)),
            "away_strikeouts": float(key_stats.get("away", {}).get("kk", 0)),
            "away_doubleplays": float(key_stats.get("away", {}).get("gd", 0)),
            "away_errors": float(key_stats.get("away", {}).get("err", 0)),
            "home_pitching": float(sum(float(i["era"]) for i in pitchers_boxscore.get("home", []))),
            "away_pitching": float(sum(float(i["era"]) for i in pitchers_boxscore.get("away", []))),
            "win": 1 if score_board.get("home", {}).get("r", 0) > score_board.get("away", {}).get("r", 0) else (
                2 if score_board.get("home", {}).get("r", 0) < score_board.get("away", {}).get("r", 0) else 3),
            # home win = 1, away win = 2, draw = 3,
            "home_starter": home_starter,
            "away_starter": away_starter
        }
```

```python
# progress_game_info를 호출하여 결과 수집
results = [progress_game_info(game_id) for game_id in game_ids_lst]

# None 값 제거 (만약 progress_game_info가 None을 반환할 수 있는 경우)
results = [result for result in results if result is not None]

# DataFrame으로 변환
df = pd.DataFrame(results)

# 결과 확인
print(df.head())
```

```python
# 날짜 포맷 변경 (YYYY-MM-DD 형식으로 변환)
df["game_date_time"] = pd.to_datetime(df["game_date_time"]).dt.strftime('%Y-%m-%d')
```

```python
# game_id 드랍
df = df.drop(columns=['game_id'])
```
